```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(n_samples=100, noise=0.2, random_state=1)
```

```python
X.shape
```

```python
y.shape
```

```python
x_train,x_test=X[:30,:],X[30:,:]
```

```python
y_train,y_test=y[:30],y[30:]
```

```python
model=Sequential()
model.add(Dense(500,activation='relu',input_dim=2))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics='accuracy')
```

```python
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=4000)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'],label='train')
plt.plot(history.history['val_accuracy'],label='test')
plt.legend()
```

```python
from keras.regularizers import l2
model_l2=Sequential()
model_l2.add(Dense(500,activation='relu',input_dim=2,kernel_regularizer=l2(0.001)))
model_l2.add(Dense(1,activation='sigmoid'))
```

```python
model_l2.compile(optimizer="adam",loss="binary_crossentropy",metrics='accuracy')
```

```python
history2=model_l2.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=4000)
```

```python
#import matplotlib.pyplot as plt
plt.plot(history2.history['accuracy'],label='train')
plt.plot(history2.history['val_accuracy'],label='test')
plt.legend()
```

```python
from keras.regularizers import l1
model_l1=Sequential()
model_l1.add(Dense(500,activation='relu',input_dim=2,kernel_regularizer=l1(0.001)))
model_l1.add(Dense(1,activation='sigmoid'))
```

```python
model_l1.compile(optimizer="adam",loss="binary_crossentropy",metrics='accuracy')
```

```python
history3=model_l1.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=4000)
```

```python
plt.plot(history3.history['accuracy'],label='train')
plt.plot(history3.history['val_accuracy'],label='test')
plt.legend()
```

```python
# so l2 is better than l1
```
