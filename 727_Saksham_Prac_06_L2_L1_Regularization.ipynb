<h1>Implementing Regularization to Avoid overfitting in binary classification</h1>

```python
from sklearn.datasets import make_moons

from keras.models import Sequential

from keras.layers import Dense
```

```python
X, y = make_moons(n_samples= 100, noise= 0.2, random_state= 1)
```

```python
X.shape
```

```python
y.shape
```

```python
# so , here I am taking 30 rows as trainning data and remaining 70 for testing.(overfitting model)

X_train, X_test = X[:30, :], X[30:, :] 
```

```python
y_train, y_test = y[:30], y[30:]
```

**Model without Regularization**

```python
model = Sequential()

model.add(Dense(500, input_dim = 2, activation = 'relu'))          #giving max Neurons on Hidden-layer to overfitt model

model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss= 'binary_crossentropy', optimizer='adam', metrics='accuracy')
```

```python
history = model.fit(X_train,y_train, validation_data=(X_test,y_test), epochs=4000)

# model will tarin, and test with data and save history
```

```python
from matplotlib import pyplot as plt

plt.plot(history.history['accuracy'], label = 'train')

plt.plot(history.history['val_accuracy'], label = 'test')

plt.legend()
```

**We can see that , Model gives best Results on Trainning data , But on testing data It's accuracy starts Decreasing.(It's Overfitted.)**

<h1>L2 Regularization</h1>

```python
from keras.regularizers import L2

model_l2 = Sequential()

model_l2.add(Dense(500, input_dim = 2, activation = 'relu', kernel_regularizer= L2(0.001)))

model_l2.add(Dense(1, activation='sigmoid'))

# L2(0.001) here , I am giving values of Penalty Hyper-Parameter (Alpha)
```

```python
model_l2.compile(loss= 'binary_crossentropy', optimizer = 'adam', metrics='accuracy')
```

```python
history2 = model_l2.fit(X_train,y_train, validation_data=(X_test,y_test), epochs=4000)

# model will tarin, and test with data and save history
```

```python
from matplotlib import pyplot as plt

plt.plot(history2.history['accuracy'], label = 'train')

plt.plot(history2.history['val_accuracy'], label = 'test')

plt.legend()
```

**Problem of Overfitting is solved, as we can see Testing data accuracy is stable (Not decreasing after epochs)**

<h1>L1 Regularization</h1>

```python
from keras.regularizers import L1

model_l1 = Sequential()

model_l1.add(Dense(500, input_dim = 2, activation = 'relu', kernel_regularizer= L1(0.0001)))

model_l1.add(Dense(1, activation='sigmoid'))

# L2(0.0001) here , I am giving values of Penalty Hyper-Parameter (Alpha)
```

```python
model_l1.compile(loss= 'binary_crossentropy', optimizer = 'adam', metrics='accuracy')
```

```python
history1 = model_l1.fit(X_train,y_train, validation_data=(X_test,y_test), epochs=2000)

# model will tarin, and test with data and save history
```

```python
from matplotlib import pyplot as plt

plt.plot(history1.history['accuracy'], label = 'train')

plt.plot(history1.history['val_accuracy'], label = 'test')

plt.legend()
```
