# Multiclass Classification

```python
import pandas as pd
from keras.models import Sequential# supervised
from keras.layers import Dense
from sklearn.datasets import load_iris
import numpy as np
```

```python
data = load_iris()
```

```python
df = pd.DataFrame(data.data,columns = data.feature_names)
df2 = pd.DataFrame(data.target,columns = ['Target'])
```

```python
df2.head()
```

```python
df.head()
```

```python
from keras.utils import np_utils
encoded_Y=np_utils.to_categorical(df2)
encoded_Y
```

```python
model=Sequential()
model.add(Dense(8,input_dim=4,activation='relu')) 
model.add(Dense(6,activation='relu'))
model.add(Dense(3,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics='accuracy') 
```

```python
model.fit(df,encoded_Y,epochs=100,batch_size=10)
```

```python
predictions=model.predict(df)
```

```python
predictions
```

```python
for i in range(35,130,3):
    print(predictions[i],encoded_Y[i])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df,encoded_Y,random_state=1,test_size=0.25)
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model=Sequential()
model.add(Dense(68,input_dim=4,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(3,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics='accuracy')
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=10)
```

```python
y_pred = model.predict(x_test)

y_pred

#y_pred = model2.predict(x_test)
```

```python
def vec2scalar(list_):
    pred = []
    for i in list_:
        val = [round(j,0) for j in i]
        if val[0]==1.0:
            pred.append(0)
        elif val[1]==1.0:
            pred.append(1)
        else:
            pred.append(2)
    return pred
```

```python
print(vec2scalar(y_pred))
```
