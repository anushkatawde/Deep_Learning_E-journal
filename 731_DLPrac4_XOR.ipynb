```python
from keras.models import Sequential# supervised
from keras.layers import Dense
import numpy as np
```

```python
model=Sequential()
model.add(Dense(units=2,input_dim=2,activation='relu')) 
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics='accuracy')
print(model.summary())
print(model.get_weights())
```

```python
X=np.array([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
print(X)
```

```python
y = np.array([0.,1.,1.,0.])
print(y)
```

```python
model.fit(X,y,epochs=400,batch_size=4)
```

```python
print(model.get_weights())
```

```python
y_pred=model.predict(X)
y_pred
```

```python
def prob_Bin(list_):
    pred = []
    for i in list_:
        if i>=0.5:
            pred.append(1)
        else:
            pred.append(0)
    return pred
```

```python
prob_Bin(y_pred)
```

```python
_, accuracy = model.evaluate(X, y)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

# Perform multiclass classification on the wine data set using deep feed forward neural network

```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
data= load_wine()
x = pd.DataFrame(data.data,columns = data.feature_names)
y = pd.DataFrame(data.target,columns = ['Target'])
x
```

```python
y
```

```python
from keras.utils import np_utils
encoded_Y=np_utils.to_categorical(y)
encoded_Y
```

```python
model=Sequential()
model.add(Dense(12,input_dim=13,activation='relu')) 
model.add(Dense(8,activation='relu'))
model.add(Dense(3,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics='accuracy')
```

```python
model.fit(x,encoded_Y,epochs=100,batch_size=10)
```

```python
predictions=model.predict(x)
```

```python
predictions
```

```python
def vec2scalar(list_):
    pred = []
    for i in list_:
        val = [round(j,0) for j in i]
        if val[0]==1.0:
            pred.append(0)
        elif val[1]==1.0:
            pred.append(1)
        else:
            pred.append(2)
    return pred


print(vec2scalar(predictions))
```

```python
print(vec2scalar(predictions))
```

```python
_, accuracy = model.evaluate(x, encoded_Y)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
