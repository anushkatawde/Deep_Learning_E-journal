# **Create a Feed Forward Neural Network to predict Diabetes**

```python
import pandas as pd
df=pd.read_csv("pima-indians-diabetes.csv")
df.head(3)
```

```python
df.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
class_nn=Sequential()
```

```python
class_nn.add(Dense(units=10,activation='relu',input_dim=8))
class_nn.add(Dense(units=8,activation='relu'))
class_nn.add(Dense(units=1,activation='sigmoid'))
```

```python
class_nn.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X.head(2)
```

```python
y.head(2)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=1,test_size=0.25)
```

```python
x_train.shape
```

```python
class_nn.fit(x_train,y_train,epochs=500,batch_size=20) # 576/20
```

```python
pred=class_nn.predict(x_test)
```

```python
pred.shape
```

```python
class_label=[] # sigmoid use kiya toh value 1 nad 2 ke bich wala aaya usse binay banaya
for i in range(192):
    if(pred[i]>0.5):
        class_label.append(1)
    else:
        class_label.append(0)
```

```python
class_label
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,class_label))
```
